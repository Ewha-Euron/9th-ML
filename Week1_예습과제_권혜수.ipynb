# **Numpy** (ndarrry)

- ndarray 타입의 데이터만 다룬다
- 내부 원소는 동일한 dtype으로만 저장




```python
import numpy as np
```

```python
##array() 함수 : 다양한 인자를 입력받아 ndarray로 변환
##np.array()는 하나의 인자만 받음
array1 = np.array([1,2,3])
print('array1 type: ', type(array1))
print('array1 array 형태: ', array1.shape) ##shape속성 -> 행, 열의 수 -> 차원의 수

array2 = np.array([[1,2,3],
                   [2,3,4]])
print('array2 type: ', type(array2))
print('array2 array 형태: ', array2.shape)

array3 = np.array([[1,2,3]])
print('array3 type: ', type(array3))
print('array3 array 형태: ', array3.shape)
```

```python
##ndarray.ndim 속으로 차원 알기

print('array1: {0}차원, array2: {1}차원, array3: {2}차원'.format(array1.ndim, array2.ndim, array3.ndim)) ##format안에서 몇번째 가져올지
```

```python
##ndarray의 데이터 타입 -> dtype 속성

list1 = [1,2,3]
print(type(list1))
array1 = np.array(list1)
print(type(array1))
print(array1, array1.dtype)
```

```python
##데이터 타입이 섞였으면?

list2 = [1,2,'test']
array2 = np.array(list2)
print(array2, array2.dtype)

list3 = [1, 2, 3.0]
array3 = np. array(list3)
print(array3, array3.dtype)
```

```python
##astrype('') 데이터 타입 변경

#int ndarray 생성-> float64
array_int = np.array([1,2,3])
array_float = array_int.astype('float64')
print(array_float, array_float.dtype)

#float -> int32
array_int1 = array_float.astype('int32')
print(array_int1, array_int1.dtype)

#float 생성 -> int32
array_float1 = np.array([1.1, 2.1, 3.1])
array_int1 = array_float1.astype('int32')
print(array_int1, array_int1.dtype)
```

```python
##andarray 편하게 생성
#np.arange()
sequence_array = np.arange(10)
print(sequence_array)
print(sequence_array.dtype, sequence_array.shape)

#zeros(), ones() -> dtype 디폴트 float64
zeroe_array = np.zeros((3,2), dtype = 'int32')
print(zeroe_array)
print(zeroe_array.dtype, zeroe_array.shape)

ones_array = np.ones((3,2))
print(ones_array)
print(ones_array.dtype, ones_array.shape)
```

```python
#reshape
# 0〜9까지의 1차원 ndarray를 2로우x5칼럼과 5로우x2칼럼 형태로 2차원 ndarray로 변환

array1 = np.arange(10)
print('array1:\n', array1)
array2 = array1.reshape(2,5)
print('array2:\n', array2)
array3 = array1.reshape(5,2)
print('array3:\n', array3)
```

```python
array1.reshape(3,4) ##수가 맞아야 함
```

```python
# reshape()에  -1값을 인자로 적용한 경우에 어떻게 ndarray의 size를 변경?

array1 = np.arange(10)
print(array1)
array2 = array1.reshape(-1, 5)
print('array2 shape', array2.shape)
array3 = array1.reshape(5, -1)
print('array3 shape', array3.shape)
```

```python
array1 = np.arange(10)
array4 = array1.reshape(-1, 4) ## 수가 맞아야 함
```

```python
## reshape(-1,1)을 이용해 3차원을 2차원으로, 1차원을 2차원으로 변경

array1 = np.arange(8)
array3d = array1.reshape(2,2,2)
print(array3d.tolist())

#3차원 -> 2차원
array2d = array3d.reshape(-1,1)
print(array2d.tolist())
print("크기는 ", array2d.shape)

#1차원 -> 2차원
array2d_1 = array1.reshape(-1,1)
print(array2d_1.tolist())
print("크기는 ", array2d_1.shape)
```

```python
# 1부터 9 까지의 1차원 ndarray 생성
array1 = np.arange(10)
print('array1: ', array1)
#index는 0부터 시작
value = array1[2]
print('value: ', value)
print(type(value))
#맨뒤의 값
print('맨뒤의 값: ', array1[-1], ', 뒤에서 두번째 값: ', array1[-2])
```

```python
#정렬하자
#sort()의 두가지 방식

#넘파이에서 sort() 호출
org_array = np.array([1, 9, 8, 2])
print("원본 행렬: ", org_array)

sort_array1 = np.sort(org_array)
print("넘파이 sort() 호출 후 반환된 행렬 ", sort_array1)
print("원본 행렬: ", org_array)

#ndarray에서 sort() 호출
sort_array2 = org_array.sort()
print(".sort() 호출 후 반환된 행렬: ", sort_array2)
print("원본 행렬 ", org_array)
```

```python
#내림차순으로 정렬
sort_array1_desc = np.sort(org_array[::-1])
print("내림차순으로 정렬: ", sort_array1_desc)
```

```python
#2차원
array2d = np.array([[8,12], [7,1 ]])

#로우 방향으로 정렬
sort_array2d_axis0 = np.sort(array2d, axis=0)
print("로우 방향으로 정렬:\n", sort_array2d_axis0)

#칼럼 방향으로 정렬
sort_array2d_axis1 = np.sort(array2d, axis=1)
print("칼럼 방향으로 정렬:\n", sort_array2d_axis1)
```

```python
#정렬된 행렬의 인덱스 반환하기
#np.argsort()
org_array = np.array([3, 1, 9, 5])
sort_indices = np.argsort(org_array)
print(type(sort_indices))
print(sort_indices)
```

```python
#내림차순 정렬시 인덱스 반환하기
org_array = np.array([3, 1, 9, 5])
sort_indices_decs = np.argsort(org_array)[::-1]
print(sort_indices_decs)
```

```python
#이름 & 성적

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array= np.array([78, 95, 84, 98, 88])

#성적 순으로 정렬된 이름
sort_indices_asc = np.argsort(score_array)
print("성적 오름차순 시 score_array의 인덱스: ", sort_indices_asc)
print("성적 오름차순으로 이름 출력: ", name_array[sort_indices_asc])
```

```python
#선형대수 연산

A = np.array([[1, 2, 3], [4, 5, 6]])
B  = np.array([[7, 8], [9, 10], [11, 12]])

#행렬 곱 np.dot()
dot_product = np.dot(A, B)
print("곱셉:\n", dot_product)

#전치 행렬 np.transpose()

transpose_mat = np.transpose(A)
print("전치:\n", transpose_mat)
```

```python
#if else 식으로 새로운 'Age_Cat'칼럼 나이가 15세 이하면 Child, 15~60세 사이는 Adult, 61세 이상 Elderly
titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x: 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
titanic_df['Age_Cat'].value_counts()
```

```python
#나이에 따라 세분화하는 함수

def get_category(age):
	cat = ""

	if age <= 5: cat = 'Baby'
	elif age <= 12 : cat = "Child"
	elif age <= 18 : cat = 'Teenager'
	elif age <= 25: cat = 'Student'
	elif age <= 35: cat = 'Young Adult'
	elif age <= 60: cat = 'Adult1'
	else :   cat = 'Elderly'

	return cat

#칼럼 만들기
titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_Cat']].head()
```

astype() 메서드: ndarray 내 데이터값의 타입 변경


```python
array_int = np.array([1,2,3])
array_float = array_int.astype('float64')
print(array_float, array_float.dtype)

array_int1 = array_float.astype('int32')
print(array_int1, array_int1.dtype)

array_float1 = np.array([1.1, 2.1, 3.1])
array_int2 = array_float1.astype('int32')
print(array_int2, array_int2.dtype)
```

**arange()**: array를 range()로 표현

**reshape()** : ndarray를 특정 차원 및 크기로 변환

```python
#reshape()
array1 = np.arange(10)
print('array1:\n', array1)

array2= array1.reshape(2,5)
print('array2:\n', array2)

array3 = array1.reshape(5,2)
print('array3:\n', array3)
```

```python
#크기가 맞지 않으면 오류가 발생함
array1.reshape(4,3)
```

```python
#인자를 -1
array1 = np.arange(10)
print(array1)

array2 = array1.reshape(-1,5) #5column에 맞게 row를 자동으로 생성해 변환
print('array2 shape: ', array2.shape)

array3 = array1.reshape(5,-1) #5row에 맞게 column을 자동으로 생성해 변환
print('array3 shape:', array3.shape)
```

```python
array1 = np.arrange(10)
array4 =array1.reshape(-1,4) #column이 4인 row를 자동 생성할 수 없기에 오류 발생
```

reshape(-1,1)와 같은 형태로 자주 사용됨, =>ndarray를 반드시 1개의 칼럼을 가진 2차원 형태로 변환

**Fancy Indexing**
:list나 ndarray로 인덱스 집합을 지정하면 해당 위치의 인덱스 ndarray를 반환


```python
array1d = np.arange(start = 1, stop = 10)
array2d = array1d.reshape(3,3)

array3 = array2d[[0,1], 2] #row 축에 fancy indexing. (row, col)인덱스가 (0,2), (1,2)로 적용됨
print('array2d[[0,1], 2]=>', array3.tolist())

array4 = array2d[[0,1], 0:2] #((0,0), (0,1)), ((1,0), (1,1)) 인덱싱
print('array2d[[0,1], 0:2]=>', array4.tolist())


array5 = array2d[[0,1]] #((0:), (1:))
print('array2d[[0,1]]=>', array5.tolist())
```

**Boolean Indexing**

**행렬의 정렬 - sort(), argsort()**

**sort()**
*   행렬을 정렬
*   np.sort() : 원 행렬 유지한 채, 정렬된 행렬을 반환
*   ndarray.sort() : 원 행렬 자체를 정렬. 반환값은 None


**argsort()**
*    정렬된 행렬의 인덱스를 반환
*   항목 추가



```python
org_array = np.array([3,1,9,5])
print('원본 행렬: ', org_array)
#np.sort()로 정렬
sort_array1 = np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 행렬:', sort_array1)
```

```python
print('np.sort() 호출 후 원본 행렬:', org_array)
#ndarray.sort()로 정렬
sort_array2 = org_array.sort()
print('org_array.sort() 호출 후 반환된 행렬:', sort_array2)
print('org_array.sort() 호출 후 원본 행렬:', org_array)
```

```python
#내림차순으로 정렬
sort_array1_desc = np.sort(org_array)[::-1]
print('내림차순으로 정렬:', sort_array1_desc)
```

```python
#내림차순 시 argsort()
org_array = np.array([3,1,9,5])
sort_indices_desc = np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 행렬의 인덱스:', sort_indices_desc)
```

```python
#argsort()의 활용
#numpy의 ndarray는 RDBMS의 TABLE 칼럼이나 Pandas의 DataFrame 갈럼과 같은 메타 데이터를 가질 수 없음. 실제값-메타데이터를 별도의 ndarray로 각각 가짐
import numpy as np

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array = np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스:', sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:', name_array[sort_indices_asc])
```

**DataFrame과 ndarray 상호 변환**


1.   ndarray, 리스트, 딕셔너리를 DataFrame으로 변환


*   DataFrame은 리스트와 ndarry와 달린 칼럼명을 가짐
*   DataFrame은 2차원 데이터->2차원 이하의 데이터들만 변환가능

```python
#1치원 리스트를 DataFrame으로 변환
import numpy as np

col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

#list를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

#ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)

#1차원 형태의 데이터를 기반으로 DataFrame을 생성하므로 칼럼명이 한개만 필요
```

```python
#2차원 형태의 데이터
#3개의 칼럼명이 필요
col_name2 = ['col1', 'col2', 'col3']

#2행 3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3], [11,12,13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape)
df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns = col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

**apply lambda 식으로 데이터 가공**


*  속도 면에서 더 빠르거나 복잡한 데이터 가공 필요 시 apply lambda 사용


```python
#lambda 함수 사용
def get_square(a):
  return a**2
print('3의 제곱은:', get_square(3))
```

```python
#위 함수를 lambda함수로 바로 변환
lambda_square = lambda x: x**2
print('3의 제곱은:', lambda_square(3))
```

```python
#여러 개의 인자값-map()함수 결합
a = [1,2,3]
squares = map(lambda x: x**2, a)
list(squares)
```

```python
#apply lambda
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x: len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x <=15 else 'Adult') #반환 값이 if else문보다 먼저 나옴
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
#else if는 적용X. 대신 else절을 내포해 if else 적용
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
#나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
  cat = ''
  if age <=5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <=18: cat = 'Teenager'
  elif age<=25 : cat = 'Student'
  elif age <=35: cat= 'Young Adult'
  elif age <=60: cat = 'Adult'
  else: cat = 'Elderly'

  return cat

#lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정.
#get_catagory(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```

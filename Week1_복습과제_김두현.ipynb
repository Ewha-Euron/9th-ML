### 21. Given a DataFrame of random numeric values:

````python
# ÎûúÎç§ ÏãúÎìú Í≥†Ï†ï
np.random.seed(2025)

df = pd.DataFrame(np.random.random(size=(5, 3)))


````

````python
print(df)
row_mean = df.mean(axis=1)
print(row_mean)
subtract = df.sub(row_mean, axis=0)
print(subtract)
````

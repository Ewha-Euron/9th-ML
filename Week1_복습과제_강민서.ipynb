### 21. Given a DataFrame of random numeric values:

````python
# ÎûúÎç§ ÏãúÎìú Í≥†Ï†ï
np.random.seed(2025)

df = pd.DataFrame(np.random.random(size=(5, 3)))

df2 = df.sub(df.mean(axis=1), axis=0)

print(df)
print(df2)
````

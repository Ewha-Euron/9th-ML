```python
train['playerID_len'] = train['playerID'].apply(lambda x : len(x))
train[['playerID','playerID_len']].head(5)
```

```python
train['child_adult'] = train['yearID'].apply(lambda x : 'adult' if x<2000 else 'child')
train[['yearID','child_adult']].head(7)
```

```python
train['Age_cat'] = train['yearID'].apply(lambda x : 'child' if x>=2000 else('semiadult'if x>=1980 else 'adult'))
train['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(yearID):
  cat = ''
  if yearID >=2010: cat = 'student'
  elif yearID >= 2000: cat = 'beginner'
  elif yearID >= 1980: cat = 'expert'
  elif yearID >= 1950: cat = 'adult'
  elif yearID >=1900: cat = 'elderly'
  else: cat = 'forefather'

  return cat

train['yearID_cat'] = train['yearID'].apply(lambda x : get_category(x))
train[['yearID','yearID_cat']].head()
```

### ë¶“ê½ƒ í’ˆì¢… ì˜ˆì¸¡í•˜ê¸°

```python
# DFë¡œ ë³€í™˜í•´ í”¼ì²˜ì™€ ë°ì´í„° ê°’ êµ¬ì„± ì•Œì•„ë³´ê¸°
import pandas as pd

# ë¶“ê½ƒ ë°ì´í„°ì„¸íŠ¸ ë¡œë”©
iris = load_iris() 

# iris.data: iris ë°ì´í„° ì„¸íŠ¸ì—ì„œ í”¼ì²˜ë§Œìœ¼ë¡œ êµ¬ì„±ëœ ë°ì´í„°ë¥¼ numpyë¡œ ê°€ì§„ë‹¤
# ê½ƒìì˜ ê¸¸ì´ì™€ ë„ˆë¹„, ê½ƒë°›ì¹¨ì˜ ë„ˆë¹„ì™€ ê¸¸ì´ í”¼ì²˜(ì…ë ¥ë³€ìˆ˜)ë¥¼ numpy ë°°ì—´ í˜•íƒœë¡œ ì €ì¥í–ˆë‹¤ëŠ” ëœ»
iris_data = iris.data

# iris.target: ë¶“ê½ƒ ë°ì´í„° ì„¸íŠ¸ì—ì„œ label(ê° ë°ì´í„°ì˜ ê²°ì •ê°’) ë°ì´í„°ë¥¼ numpyë¡œ ê°€ì§„ë‹¤
# 0, 1, 2 ì„¸ ê°€ì§€ ìˆ«ìë¡œ ë˜ì–´ ìˆê³ , ê°ê°ì´ ê½ƒ ì¢…ë¥˜(Setosa, Versicolor, Virginica)ë¥¼ ì˜ë¯¸
iris_label = iris.target
print('iris target ê°’:', iris_label)
print('iris target ëª…:', iris.target_names)

# ë¶“ê½ƒ ë°ì´í„°ì…‹ì„ DataFrameìœ¼ë¡œ ë³€í™˜
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['label']=iris.target
iris_df.head(3)
```

iris: ì‚¬ì´í‚·ëŸ°ì´ ì œê³µí•˜ëŠ” Bunch ê°ì²´ <br>
load_iris()ê°€ ë°˜í™˜í•˜ëŠ” ì´ Bunch ì•ˆì—ëŠ” ì—¬ëŸ¬ ê°€ì§€ ì†ì„±ì´ í¬í•¨ë¨. 

- iris.data â†’ í”¼ì²˜ê°’ (numpy ë°°ì—´)
- iris.target â†’ ì •ë‹µê°’ (numpy ë°°ì—´)

- iris.DESCR â†’ ë°ì´í„°ì…‹ ì„¤ëª… í…ìŠ¤íŠ¸

--- 
iris.target = ë„˜íŒŒì´ ë°°ì—´ (ì •ë‹µ ê°’)
iris_label = ìœ„ ë°°ì—´ì„ ë³µì‚¬í•œ ë³€ìˆ˜
iris.target_names = ì •ë‹µ ê°’(0/1/2)ì— ëŒ€ì‘í•˜ëŠ” í’ˆì¢… ì´ë¦„í‘œ

```python
# í•™ìŠµìš©, í…ŒìŠ¤íŠ¸ìš© ë°ì´í„° ë¶„ë¦¬í•˜ê¸°
# train_test_split() API í™œìš©: ê°ê°ì˜ ë°ì´í„°ë¥¼ test_size íŒŒë¼ë¯¸í„° ì…ë ¥ê°’ì˜ ë¹„ìœ¨ë¡œ ë¶„í• í•¨ 
# test_size=0.2 ë¼ê³  ì„¤ì •í•˜ë©´ ì „ì²´ ë°ì´í„° ì¤‘ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ 20%, í•™ìŠµ ë°ì´í„°ë¥¼ 80%ë¡œ ë°ì´í„°ë¥¼ ë¶„í• í•¨

X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
```

- ì²« ë²ˆì§¸ ì…ë ¥ íŒŒë¼ë¯¸í„° iris_data: **í”¼ì²˜ ë°ì´í„° ì„¸íŠ¸**  (X)
- ë‘ ë²ˆì§¸ ì…ë ¥ íŒŒë¼ë¯¸í„° iris_label: **label ë°ì´í„° ì„¸íŠ¸** (y)
- test_sizeëŠ” ìœ„ ì£¼ì„ì˜ ë‚´ìš©ê³¼ ê°™ìŒ. ì—¬ê¸°ì„œëŠ” í…ŒìŠ¤íŠ¸ ë°ì´í„° 20% í™œìš©í•œë‹¨ ëœ».
- ë„¤ ë²ˆì§¸ ì…ë ¥ íŒŒë¼ë¯¸í„° random_state=11: í˜¸ì¶œ ì‹œë§ˆë‹¤ ê°™ì€ í•™ìŠµ/í…ŒìŠ¤íŠ¸ìš© ë°ì´í„°ì…‹ì„ ìƒì„±í•˜ê¸° ìœ„í•´ ì£¼ì–´ì§€ëŠ” ë‚œìˆ˜ ë°œìƒ ê°’. ì´ë¥¼ ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ train_test_split ìˆ˜í–‰í•  ë•Œë§ˆë‹¤ ë‹¤ë¥¸ ë°ì´í„°ë¥¼ ìƒì„±í•˜ê²Œ ë¨. (ìˆ«ìëŠ” ì•„ë¬´ ê°’ ì§€ì •í•´ë„ ë¨.)

```python
# ì˜ì‚¬ê²°ì •íŠ¸ë¦¬ ì´ìš©í•´ ë°ì´í„°ì— ëŒ€í•œ í•™ìŠµê³¼ ì˜ˆì¸¡ ìˆ˜í–‰í•˜ê¸° 
# ì‚¬ì´í‚·ëŸ°ì˜ ì˜ì‚¬ê²°ì •íŠ¸ë¦¬ í´ë˜ìŠ¤ë¥¼ ê°ì²´ë¡œ ìƒì„±
dt_clf = DecisionTreeClassifier(random_state=11) 

# í•™ìŠµ ìˆ˜í–‰
# fit() ë©”ì„œë“œì— í•™ìŠµìš© í”¼ì²˜ ë°ì´í„° ì†ì„±, ê²°ì •ê°’ ë°ì´í„°ì…‹ ì…ë ¥í•´ í˜¸ì¶œí•˜ë©´ í•™ìŠµ ìˆ˜í–‰í•¨.
dt_clf.fit(X_train, y_train)
```

ğŸ‘†ğŸ» í•™ìŠµ ë°ì´í„°(X_train, y_train)ë¥¼ ë°”íƒ•ìœ¼ë¡œ í•™ìŠµì„ ì™„ë£Œí•œ ìƒí™©. ì´ë ‡ê²Œ í•™ìŠµëœ DecisionTreeClassifier ê°ì²´ë¥¼ ì´ìš©í•´ **ì˜ˆì¸¡** ì„ ìˆ˜í–‰í•œë‹¤. <br>
ì˜ˆì¸¡: ë°˜ë“œì‹œ í•™ìŠµë°ì´í„°ê°€ ì•„ë‹Œ ë‹¤ë¥¸ ë°ì´í„°!! ë¥¼ ì´ìš©í•´ì•¼ í•˜ë©°, ì¼ë°˜ì ìœ¼ë¡œ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ë¥¼ í™œìš©í•œë‹¤. <br>
ì˜ì‚¬ê²°ì •íŠ¸ë¦¬ ê°ì²´ì˜ `predict()` ë©”ì„œë“œì— ì´ ë°ì´í„° ì„¸íŠ¸ë¥¼ ì…ë ¥í•´ í˜¸ì¶œí•˜ë©´, í•™ìŠµëœ ëª¨ë¸ ê¸°ë°˜ì—ì„œ í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì— ëŒ€í•œ ì˜ˆì¸¡ê°’ì„ ë°˜í™˜í•œë‹¤.

```python
# í•™ìŠµ ì™„ë£Œëœ ì˜ì‚¬ê²°ì •íŠ¸ë¦¬ ê°ì²´ì—ì„œ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰í•˜ê¸°
pred = dt_clf.predict(X_test)
```

```python
# DTCì˜ ì˜ˆì¸¡ ì„±ëŠ¥ í‰ê°€í•˜ê¸°: ì •í™•ë„ accuracy_score() 
# ì˜ˆì¸¡í•œ ë¶“ê½ƒ í’ˆì¢…ê³¼ ì‹¤ì œ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ì˜ ë¶“ê½ƒ í’ˆì¢…ì´ ì¼ì¹˜í•˜ëŠ” ì •ë„

from sklearn.metrics import accuracy_score
print('ì˜ˆì¸¡ ì •í™•ë„: {0:.4f}'.format(accuracy_score(y_test, pred)))
```

X ê³„ì—´ = ì…ë ¥ê°’(í”¼ì²˜) <br>
y ê³„ì—´ = ì •ë‹µê°’(ë ˆì´ë¸”) <br>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

dt_clf = DecisionTreeClassifier() 
iris_data = load_iris()

# test_train_split(í”¼ì²˜ ë°ì´í„°ì…‹, ë ˆì´ë¸” ë°ì´í„°ì…‹)
# í…ŒìŠ¤íŠ¸ ë°ì´í„°ì…‹ì„ 30%ë¡œ, random_stateë¥¼ 121ë¡œ
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, test_size=0.3, random_state=121)

# dataëŠ” ê° í”¼ì²˜ì˜ ìˆ˜ì¹˜(ê½ƒì ê¸¸ì´.. etc)ê³  targetì€ 0, 1, 2 ê°’ 
```

```python
# ìœ„ì˜ í•™ìŠµ ë°ì´í„°ë¥¼ ê¸°ë°˜ìœ¼ë¡œ DTCë¥¼ í•™ìŠµí•˜ê³  (fit ë©”ì„œë“œ) ì´ ëª¨ë¸ì„ ì´ìš©í•´ ì˜ˆì¸¡ ì •í™•ë„ë¥¼ ì¸¡ì •í•˜ê¸°

# í•™ìŠµìš© ì…ë ¥ ë°ì´í„°ì™€ ì •ë‹µì„ ì´ìš©í•´ ê·œì¹™ì„ í•™ìŠµí•¨
dt_clf.fit(X_train, y_train)

# í•™ìŠµì´ ì™„ë£Œëœ ëª¨ë¸ì— í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ë¥¼ ë„£ìŒ - 0, 1, 2ë¡œ êµ¬ì„±ëœ ë°°ì—´ì´ predì— ì €ì¥ë¨
pred = dt_clf.predict(X_test)


print('ì˜ˆì¸¡ ì •í™•ë„: {0:.4f}'.format(accuracy_score(y_test, pred)))
```

dt_clf.fit(X_train, y_train)ì—ì„œëŠ” **y_train**ì´ ì •ë‹µ ì—­í• ì„ í•œë‹¤. (êµê³¼ì„œ ëŠë‚Œ) <br>
accuracy_score(y_test, pred)ì—ì„œëŠ” X_testì— ëŒ€í•´ ëª¨ë¸ì´ ë‚¸ ì˜ˆì¸¡(pred)ì„ ì‹¤ì œ ì •ë‹µ(**y_test**)ê³¼ ë¹„êµí•´ì„œ ë§ì·„ëŠ”ì§€ í™•ì¸í•œë‹¤. 

---
y_train = í•™ìŠµ ë‹¨ê³„ì—ì„œ ì“°ëŠ” ì •ë‹µ (ëª¨ë¸ì´ ë°°ìš¸ ë•Œ ì°¸ê³ ) <br>
y_test = í‰ê°€ ë‹¨ê³„ì—ì„œ ì“°ëŠ” ì •ë‹µ (ëª¨ë¸ì´ ì œëŒ€ë¡œ ë°°ì› ëŠ”ì§€ ì±„ì í•  ë•Œ ì‚¬ìš©)

#### êµì°¨ ê²€ì¦

#### K í´ë“œ êµì°¨ ê²€ì¦

```python
# 1. K í´ë“œ êµì°¨ ê²€ì¦: KFold í´ë˜ìŠ¤ 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

# K=5ì¸ í´ë“œ ì„¸íŠ¸ë¡œ ë¶„ë¦¬í•˜ëŠ” KFold ê°ì²´ì™€ í´ë“œ ì„¸íŠ¸ë³„ ì •í™•ë„ë¥¼ ë‹´ì„ ë¦¬ìŠ¤íŠ¸ ê°ì²´ ìƒì„±
kfold = KFold(n_splits=5)
cv_accuracy = []
print('ë¶“ê½ƒ ë°ì´í„° ì„¸íŠ¸ í¬ê¸°:', features.shape[0])
```

```python
# ìƒì„±ëœ KFold ê°ì²´ì— split()ì„ í˜¸ì¶œí•´ ì „ì²´ ë¶“ê½ƒ ë°ì´í„°ë¥¼ 5ê°œì˜ ë°ì´í„° í´ë“œ ì„¸íŠ¸ë¡œ ë¶„ë¦¬ 
# 150ê°œ ì¤‘ 4/5ì¸ 120ê°œë¥¼ í•™ìŠµ, 1/5ì¸ 30ê°œë¥¼ ê²€ì¦ í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ë¡œ í™œìš©
# split()ì„ í˜¸ì¶œí•´ êµì°¨ ê²€ì¦ ìˆ˜í–‰ ì‹œë§ˆë‹¤ í•™ìŠµ/ê²€ì¦ì„ ë°˜ë³µí•´ ì˜ˆì¸¡ ì •í™•ë„ë¥¼ ì¸¡ì •

n_iter = 0

# KFold ê°ì²´ì˜ split()ì„ í˜¸ì¶œí•˜ë©´ í´ë“œë³„ í•™ìŠµìš©, ê²€ì¦ìš© í…ŒìŠ¤íŠ¸ì˜ ë¡œìš° ì¸ë±ìŠ¤ë¥¼ ë°°ì—´ë¡œ ë°˜í™˜
for train_index, test_index in kfold.split(features): 
    # kfold.split()ìœ¼ë¡œ ë°˜í™˜ëœ ì¸ë±ìŠ¤ë¥¼ ì´ìš©í•´ í•™ìŠµìš©, ê²€ì¦ìš© í…ŒìŠ¤íŠ¸ ë°ì´í„° ì¶”ì¶œ
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]

    # í•™ìŠµ ë° ì˜ˆì¸¡
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    n_iter += 1

    # ë°˜ë³µ ì‹œë§ˆë‹¤ ì •í™•ë„ ì¸¡ì •í•˜ê¸°
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('\n#{0} êµì°¨ ê²€ì¦ ì •í™•ë„: {1}, í•™ìŠµ ë°ì´í„° í¬ê¸°: {2}, ê²€ì¦ ë°ì´í„° í¬ê¸°: {3}'.format(n_iter, accuracy, train_size, test_size))
    print('#{0} ê²€ì¦ ì„¸íŠ¸ ì¸ë±ìŠ¤:{1}'.format(n_iter, test_index))
    cv_accuracy.append(accuracy) 

# ê°œë³„ ë°˜ë³µë³„ ì •í™•ë„ë¥¼ í•©í•´ í‰ê·  ì •í™•ë„ ê³„ì‚°
print('\n## í‰ê·  ê²€ì¦ ì •í™•ë„:', np.mean(cv_accuracy))
```

ê²€ì¦ ì„¸íŠ¸ì˜ ì¸ë±ìŠ¤ë¥¼ ë³´ë©´ 0-29, 30-59... ì²˜ëŸ¼ 30ê°œì”© ê°ê°ì˜ ê²€ì¦ ì„¸íŠ¸ ì¸ë±ìŠ¤ê°€ ìƒì„±ë˜ê³  ìˆìœ¼ë©°, ì´ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ê²€ì¦ ì„¸íŠ¸ê°€ ì¶”ì¶œë˜ê³  ìˆë‹¤ëŠ” ê±¸ ì•Œ ìˆ˜ ìˆë‹¤. 

#### Stratified K í´ë“œ

```python
# K í´ë“œì˜ í•œê³„ë¥¼ ë¨¼ì € í™•ì¸... 
# ë¶“ê½ƒ ë°ì´í„°ì…‹ì„ DFë¡œ ìƒì„±í•˜ê³  ë ˆì´ë¸” ê°’ì˜ ë¶„í¬ë„ í™•ì¸
import pandas as pd
iris = load_iris()
ifis_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['label']=iris.target
iris_df['label'].value_counts()
```

```python
# ìœ„ì˜ ì½”ë“œì—ì„œ ê°ê°ì˜ ë ˆì´ë¸”ì´ 50ê°œë¡œ ë™ì¼í•˜ê²Œ ì¶œë ¥ë˜ì—ˆë‹¤. 
# 3ê°œì˜ í´ë“œ ì„¸íŠ¸ë¥¼ KFoldë¡œ ìƒì„±í•˜ê³ , êµì°¨ ê²€ì¦ ì‹œë§ˆë‹¤ ê°ê° ìƒì„±ë˜ëŠ” í•™ìŠµ/ê²€ì¦ ë°ì´í„° ê°’ì˜ ë¶„í¬ë¥¼ í™•ì¸í•´ ë³´ê¸°
kfold = KFold(n_splits=3)
n_iter=0

# ê° ë°˜ë³µë§ˆë‹¤ í•™ìŠµìš© ì¸ë±ìŠ¤ / ê²€ì¦ìš© ì¸ë±ìŠ¤ë¥¼ ë°˜í™˜
for train_index, test_index in kfold.split(iris_df): 
    n_iter += 1

    # í•™ìŠµ ì„¸íŠ¸ì™€ ê²€ì¦ ì„¸íŠ¸ì˜ ë ˆì´ë¸”(ì •ë‹µ, í’ˆì¢… ë²ˆí˜¸)ì„ êº¼ë‚´ì˜´ 
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]
    print('## êµì°¨ ê²€ì¦: {0}'.format(n_iter))
    print('í•™ìŠµ ë ˆì´ë¸” ë°ì´í„° ë¶„í¬:\n', label_train.value_counts())
    print('ê²€ì¦ ë ˆì´ë¸” ë°ì´í„° ë¶„í¬:\n', label_test.value_counts())
    print('\n')
```

ì „ì²´ ë¶“ê½ƒ ë°ì´í„° 150ê°œì—ì„œ ë°ì´í„° í´ë“œ 3ê°œ ìƒì„± - 100ê°œì˜ í•™ìŠµ ë°ì´í„°ì™€ 50ê°œì˜ ê²€ì¦ ë°ì´í„° <br>
ì´ë•Œ, êµì°¨ ê²€ì¦ ì‹œë§ˆë‹¤ í•™ìŠµ ë ˆì´ë¸”ê³¼ ê²€ì¦ ë ˆì´ë¸”ì´ ì™„ì „íˆ ë‹¤ë¥¸ ê°’ìœ¼ë¡œ ì¶”ì¶œëœ ê²ƒì„ ë³¼ ìˆ˜ ìˆë‹¤. <br>
ì˜ˆë¥¼ ë“¤ì–´ êµì°¨ê²€ì¦ #1ì—ì„œ, í•™ìŠµ ì„¸íŠ¸ì— 1ê³¼ 2ë§Œ ìˆë‹¤. ì´ëŸ¬ë©´ ëª¨ë¸ì€ 0ì´ë¼ëŠ” ê²Œ ì¡´ì¬í•œë‹¤ëŠ” ê±¸ ëª¨ë¥´ê²Œ ëœë‹¤.

í•™ìŠµ ë ˆì´ë¸”ê³¼ ê²€ì¦ ë ˆì´ë¸” ë°ì´í„° ê°’ì˜ ë¶„í¬ë„ê°€ ê±°ì˜ ë™ì¼í•˜ê²Œ í• ë‹¹ë˜ì—ˆìŒì„ ì•Œ ìˆ˜ ìˆë‹¤. ì´ë ‡ê²Œ ë¶„í• ë˜ì–´ì•¼ ë ˆì´ë¸” ê°’ 0, 1, 2ë¥¼ ëª¨ë‘ í•™ìŠµí•˜ê³  ì´ì— ê¸°ë°˜í•´ ê²€ì¦ì„ ìˆ˜í–‰í•  ìˆ˜ ìˆë‹¤!!

```python
# StratifiedKFold ì´ìš©í•´ ë¶“ê½ƒ ë°ì´í„° êµì°¨ ê²€ì¦í•˜ê¸°

# ë°ì´í„° ë¶„ë¦¬
dt_clf = DecisionTreeClassifier(random_state=156)

skfold = StratifiedKFold(n_splits=3)
n_iter = 0
cv_accuracy = []

# StratifiedKFoldì˜ split()ì„ í˜¸ì¶œí•  ë• ë ˆì´ë¸” ë°ì´í„° ì„¸íŠ¸ë„ ë°˜ë“œì‹œ ë„£ì–´ ì£¼ì–´ì•¼ í•œë‹¤ (KFoldëŠ” í”¼ì²˜ ë°ì´í„°ì…‹ë§Œ ì…ë ¥)
for train_index, test_index in skfold.split(features, label):
    # split()ìœ¼ë¡œ ë°˜í™˜ëœ ì¸ë±ìŠ¤ë¥¼ ì´ìš©í•´ í•™ìŠµìš©, ê²€ì¦ìš© í…ŒìŠ¤íŠ¸ ë°ì´í„° ì¶”ì¶œ
    X_train, X_test = features[train_index], features[test_index] 
    y_train, y_test = label[train_index], label[test_index]

    # í•™ìŠµ ë° ì˜ˆì¸¡
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)

    # ë°˜ë³µë§ˆë‹¤ ì •í™•ë„ ì¸¡ì •
    n_iter += 1
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    train_size = X_train.shape[0]
    test_size = X_test.shape[0]
    print('\n#{0} êµì°¨ ê²€ì¦ ì •í™•ë„: {1}, í•™ìŠµ ë°ì´í„° í¬ê¸°: {2}, ê²€ì¦ ë°ì´í„° í¬ê¸°: {3}'
          .format(n_iter, accuracy, train_size, test_size))
    print('#{0} ê²€ì¦ ì„¸íŠ¸ ì¸ë±ìŠ¤:{1}'.format(n_iter, test_index))
    cv_accuracy.append(accuracy)

print('=======================================================================')

# êµì°¨ ê²€ì¦ë³„ ì •í™•ë„ ë° í‰ê·  ì •í™•ë„ ê³„ì‚°
print('## êµì°¨ ê²€ì¦ë³„ ì •í™•ë„:', np.round(cv_accuracy, 4))
print('## í‰ê·  ê²€ì¦ ì •í™•ë„:', np.round(np.mean(cv_accuracy), 4))
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# í•™ìŠµ ë°ì´í„°ëŠ” 0 ë¶€í„° 10ê¹Œì§€, í…ŒìŠ¤íŠ¸ ë°ì´í„°ëŠ” 0ë¶€í„° 5ê¹Œì§€ ê°’ì„ ê°€ì§€ëŠ” ë°ì´í„° ì„¸íŠ¸ë¡œ ìƒì„±
# fit(), transform()ì€ 2ì°¨ì› ì´ìƒ ë°ì´í„°ë§Œ ê°€ëŠ¥: reshape(-1, 1)ë¡œ ì°¨ì› ë³€ê²½
train_array = np.arange(0, 11).reshape(-1, 1)
test_array =  np.arange(0, 6).reshape(-1, 1)
```
